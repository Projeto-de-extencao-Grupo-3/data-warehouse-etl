```python
import os
import pandas as pd
from glob import glob
```

```python
df = pd.read_csv("./trusted/trusted_datatran.csv")

print("Dados do datatran: ", df.head())
```

```python
# Adicionando coluna mês a partir da coluna data
df["MES"] = pd.to_datetime(df["DATA"], errors="coerce").dt.month.astype("Int64")

df.to_csv("./refined/refined_datatran.csv", index=False)
```

```python
# Pegando os dados de acidentes de trânsito apenas do estado de São Paulo
df = df[df["ESTADO"] == "SP"]

df.to_csv("./refined/refined_datatran_sp.csv", index=False)
```

```python
# Pegando os dados de acidentes de trânsito do estado de São Paulo com falhas mecânicas ou elétricas como causa do acidente
df_final = df[
    (df["ESTADO"] == "SP") &
    (df["CAUSA_ACIDENTE"] == "Demais falhas mecânicas ou elétricas")
]

df_final.to_csv("./refined/refined_datatran_sp_falhas_mecanicas.csv", index=False)
```
