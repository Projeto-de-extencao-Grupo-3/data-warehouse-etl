```python
# Limpeza dos dados do feriado nacional, removendo coluna e padronizando o formato das colunas

df = pd.read_csv("./raw/raw_feriado_nacional.csv", header=None)

print("Dados do feriado nacional: ", df.head())

df = df.dropna(axis=1, how="all")

add_header = ["DATA", "DIA", "TIPO", "DESCRICAO"]

df.columns = add_header

df["DIA"] = (
    df["DIA"]
    .astype(str)
    .str.replace(r"^\s*Dia d[eo]\s+", "", regex=True, case=False)
    .str.strip()
)

df.to_csv("./trusted/trusted_feriado_nacional.csv", index=False)
```

```python
# Limpeza dos dados do feriado municipal, removendo coluna e padronizando o formato das colunas

df = pd.read_csv("./raw/raw_feriado_municipal.csv", header=None)

print("Dados do feriado municipal: ", df.head())

df = df.dropna(axis=1, how="all")

add_header = ["DATA", "DIA", "TIPO", "DESCRICAO", "ESTADO", "CODIGO_MUNICIPIO"]

df.columns = add_header

df["DIA"] = (
    df["DIA"]
    .astype(str)
    .str.replace(r"^\s*Dia d[eoa]\s+", "", regex=True, case=False)
    .str.strip()
)

df.to_csv("./trusted/trusted_feriado_municipal.csv", index=False)
```

```python
# Limpeza dos dados do feriado facultativo, removendo coluna e padronizando o formato das colunas

df = pd.read_csv("./raw/raw_feriado_facultativo.csv", header=None)

print("Dados do feriado facultativo: ", df.head())

df = df.dropna(axis=1, how="all")

add_header = ["DATA", "DIA", "TIPO", "DESCRICAO", "ESTADO", "CODIGO_MUNICIPIO"]

df.columns = add_header

df["DIA"] = (
    df["DIA"]
    .astype(str)
    .str.replace(r"^\s*Dia d[eoa]\s+", "", regex=True, case=False)
    .str.strip()
)

df.to_csv("./trusted/trusted_feriado_facultativo.csv", index=False)
```

```python
# Limpeza dos dados do feriado estadual, removendo coluna e padronizando o formato das colunas

df = pd.read_csv("./raw/raw_feriado_estadual.csv", header=None)

print("Dados do feriado estadual: ", df.head())

df = df.dropna(axis=1, how="all")

add_header = ["DATA", "DIA", "TIPO", "DESCRICAO", "ESTADO"]

df.columns = add_header

df["DIA"] = (
    df["DIA"]
    .astype(str)
    .str.replace(r"^\s*Dia d[eoa]\s+", "", regex=True, case=False)
    .str.strip()
)

df.to_csv("./trusted/trusted_feriado_estadual.csv", index=False)
```

```python
# Juntando os CSVs

df_estadual= pd.read_csv("./trusted/trusted_feriado_estadual.csv")
df_facultativo = pd.read_csv("./trusted/trusted_feriado_facultativo.csv")
df_nacional = pd.read_csv("./trusted/trusted_feriado_nacional.csv")
df_municipal = pd.read_csv("./trusted/trusted_feriado_municipal.csv")

df_final = pd.concat([df_estadual, df_facultativo, df_nacional, df_municipal], ignore_index=True)

df_final.to_csv("./trusted/trusted_feriados.csv", index=False)
```

```python

df_final["MES"] = pd.to_datetime(df_final["DATA"], errors="coerce").dt.month

df_final.to_csv("./refined/refined_feriados.csv", index=False)
```
