```python
# Limpeza dos dados do datatran, removendo coluna e padronizando o formato das colunas
df = pd.read_csv("./raw/raw_datatran.csv", sep=";", encoding="latin1")

print("Dados do datatran: ", df.head())

# Removendo colunas que não fazem sentido
df = df.drop(columns=["br", "km", "latitude", "longitude", "delegacia", "uop", "uso_solo", "feridos_leves", "feridos_graves", "ignorados"], errors="ignore") 
```

```python
# Padronizando nomeclatura das colunas
df.columns = df.columns.str.strip().str.upper()
df = df.rename(columns={
    "DATA_INVERSA": "DATA",
    "UF": "ESTADO",
})

# Padronizando o formato da coluna de data para fazer o cruzamento com o dataset feriados
df["DATA"] = pd.to_datetime(df["DATA"], errors="coerce").dt.strftime("%d/%m/%Y")


df.to_csv("./trusted/trusted_datatran.csv", header=True, index=False)
```
